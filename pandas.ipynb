```python
import numpy as np
#pandas数据结构和数据类型
#1.dataframe（多列对象）对象：包括series（一列对象），包括索引列
import pandas as pd

from demo1 import category

#创建series对象
#1.通过list列表来创建
s1 = pd.Series([1, 2, 3, 4, 5])
print(s1)
s3 = pd.Series([1, 2, 3, 4, 5], index=['a', 'b', 'c', 'd', 'e'])
print(s3)


```

```python
#2.通过字典或元组创建series对象
t = (1, 2, 3, 4, 5, 6)
print(pd.Series(t))

dst = {'A': 1, 'B': 2, 'C': 3}
s3 = pd.Series(dst)
print(s3)
```

```python
#4.使用numpy创建series对象
s1 = pd.Series(np.arange(10))
print(s1)
```

```python
#series对象属性
#1.
s1 = pd.Series(np.arange(6), index=[i for i in 'ABCDEF'])  #使用了列表推导式
s1 = pd.Series(data=[i for i in range(6)], index=[i for i in 'ABCDEF'])
print(s1)
```

```python
#2.获取series对象的值
print(s1.index)
```

```python
#3.获取series对象的值列
print(s1.values)
```

```python
#4.支持根据索引获取元素
print(s1['D'])
#5.同样可以修改
s1['D'] = 99
print(s1)
```

```python
#创建dataframe
#1.使用字典加列表创建df
df1 = {
    '日期': ['2021-08-21', '2021-08-22', '2021-08-23'],
    '温度': [25, 26, 50],
    '湿度': [81, 50, 56]
}
df1 = pd.DataFrame(data=df1, index=[i for i in 'ABC'])
print(df1)
```

```python
#2.使用列表加元组创建df,自定义索引
df2 = {
    ('2021-08-21', 25, 81),
    ('2021-08-22', 26, 50),
    ('2021-08-23', 27, 56)
}

df2 = pd.DataFrame(data=df2, columns=['日期', '温度', '湿度'], index=[i for i in 'ABC'])
print(df2)
```

```python
#3.使用numpy创建df
df3 = pd.DataFrame(data=np.random.randint(40, 100, (4, 6)), columns=['语文', '数学', '英语', '物理', '化学', '生物'],
                   index=['之之', '曦曦', '猫猫', '狗狗'])
print(df3)

score = np.random.randint(40, 100, (5, 6))
subjects = ['语文', '数学', '英语', '物理', '化学', '生物']
stu = ['同学' + str(i) for i in range(score.shape[0])]
df4 = pd.DataFrame(score, columns=subjects, index=stu)
print(df4)
```

```python
#简单案例
score = np.random.randint(40, 101, (10, 5))
subjects = ['语文', '数学', '英语', '物理', '化学']
stu = ['同学' + str(i) for i in range(score.shape[0])]
df = pd.DataFrame(data=score, columns=subjects, index=stu)
#或者通过df.columns和df.index修改dataframe
#或者通过rename函数 df.rename(columns,index,inplace)
print(df)
```

```python
#可以用rename 实现精准修改
#语法为 df.rename(columns={},index={},inplace=True)
df.rename(columns={'化学': '生物'})
```

```python
#dataframe对象属性
#1.shape属性
print(df.shape)

#2.index属性
print(df.index)

#3.columns属性
print(df.columns)

#4.values属性
print(df.values)

#5.转置属性
print(df.T)
```

```python
#dataframe对象属性
#1.修改行索引值
stu = ['学生_' + str(i) for i in range(df.shape[0])]
df.index = stu
print(df)
```

```python
#2.重设索引
#df.reset_index(drop=false) drop默认为false,如果为true,删除原来的索引值
df.reset_index(drop=True)
```

```python
#3.head(),查看前n条数据
df.head(3)
#tail(),查看后n条数据
df.tail(3)
```

```python
#4.info(),查看df对象的信息
print(df.info())
```

```python
#5.describe(),查看df对象的描述性统计信息
print(df.describe())
```
