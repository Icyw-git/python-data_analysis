```python
import numpy as np

# 定义矩阵A
# A = np.array([[1, 2, 3, 5], [1/2, 1, 1/2, 2], [1/3, 2, 1, 1/2], [1/5, 1/2, 2, 1]])
A = np.array([[1, 2, 3, 5], [1 / 2, 1, 1 / 2, 2], [1 / 3, 2, 1, 2], [1 / 5, 1 / 2, 1 / 2, 1]])

n = A.shape[0]  # 获取A的行

# 求出最大特征值以及对应的特征向量
eig_val, eig_vec = np.linalg.eig(A)  # eig_val是特征值， eig_vec是特征向量
Max_eig = max(eig_val)  # 求特征值的最大值

CI = (Max_eig - n) / (n - 1)
RI = [0, 0.0001, 0.52, 0.89, 1.12, 1.26, 1.36, 1.41, 1.46, 1.49, 1.52, 1.54, 1.56, 1.58, 1.59]
# 注意哦，这里的RI最多支持 n = 15
# 这里n=2时，一定是一致矩阵，所以CI = 0，我们为了避免分母为0，将这里的第二个元素改为了很接近0的正数

CR = CI / RI[n]

print('一致性指标CI=', CI)
print('一致性比例CR=', CR)

if CR < 0.10:
    print('因为CR<0.10，所以该判断矩阵A的一致性可以接受!')
else:
    print('注意：CR >= 0.10，因此该判断矩阵A需要进行修改!')
```

```python
# %% [markdown]
# # 统计图绘制示例
# 本Notebook演示如何使用Matplotlib和Seaborn绘制常见统计图表
# 按Shift+Enter逐单元格运行

# %% [markdown]
# ## 1. 导入必要的库
# 首先导入绘图和数据处理所需的各种# %%
# 基础库
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 设置图表样式
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")  # 设置调色板

# 设置图表尺寸
plt.rcParams['figure.figsize'] = (10, 6)  # 默认图表大小
plt.rcParams['figure.dpi'] = 100  # 图表分辨率

print("✅ 所有库已成功导入！")
print(f"Numpy版本: {np.__version__}")
print(f"Pandas版本: {pd.__version__}")
print(f"Matplotlib版本: {plt.matplotlib.__version__}")

plt.rcParams['font.sans-serif'] = ['SimHei', 'Microsoft YaHei', 'DejaVu Sans']
plt.rcParams['axes.unicode_minus'] = False

# %% [markdown]
# ## 2. 生成示例数据
# 创建一些模拟数据用于可视化

# %%
# 设置随机种子，确保结果可重复
np.random.seed(42)

# 生成示例数据
n_samples = 200

# 创建DataFrame
data = pd.DataFrame({
    '产品类别': np.random.choice(['A', 'B', 'C', 'D'], n_samples),
    '销售额': np.random.normal(1000, 200, n_samples).clip(500, 1500),  # 销售额，均值为1000
    '客户评分': np.random.uniform(3.0, 5.0, n_samples),  # 评分3-5
    '广告费用': np.random.exponential(500, n_samples),  # 广告费
    '月份': np.random.choice([f'{m}月' for m in range(1, 13)], n_samples),
    '地区': np.random.choice(['华北', '华东', '华南', '华西'], n_samples),
    '销量': np.random.poisson(50, n_samples),  # 销量
    '成本': np.random.normal(600, 100, n_samples),  # 成本
})

# 计算利润
data['利润'] = data['销售额'] - data['成本']
data['利润率'] = (data['利润'] / data['销售额'] * 100).round(2)

# 添加日期列
date_range = pd.date_range('2023-01-01', periods=n_samples, freq='D')
data['日期'] = date_range[:n_samples]

# 显示数据前几行
print("📊 示例数据前5行：")
display(data.head())

print(f"\n📈 数据形状: {data.shape}")
print(f"📅 日期范围: {data['日期'].min().date()} 到 {data['日期'].max().date()}")
print("\n📋 数据统计摘要：")
display(data.describe().round(2))

# %% [markdown]
# ## 3. 折线图 - 时间趋势分析
# 用于展示数据随时间的变化趋势

# %%
# 3.1 月度销售额趋势
# 按月份聚合数据
monthly_sales = data.groupby('月份')['销售额'].mean().reindex(
    [f'{m}月' for m in range(1, 13)]
)

plt.figure(figsize=(12, 6))

# 创建折线图
plt.plot(monthly_sales.index, monthly_sales.values, marker='o', markersize=8, linewidth=2.5,
         color='#FF6B6B', markerfacecolor='white', markeredgewidth=2)

# 添加数据标签
for i, (month, sales) in enumerate(zip(monthly_sales.index, monthly_sales.values)):
    plt.text(i, sales + 20, f'¥{sales:.0f}',
             ha='center', va='bottom', fontsize=9, fontweight='bold')

# 设置图表属性
plt.title('📈 月度平均销售额趋势', fontsize=16, fontweight='bold', pad=20)
plt.xlabel('月份', fontsize=12)
plt.ylabel('平均销售额（元）', fontsize=12)
plt.xticks(rotation=0)
plt.grid(True, alpha=0.3, linestyle='--')
plt.ylim(monthly_sales.min() * 0.9, monthly_sales.max() * 1.1)

# 填充区域增加视觉效果
plt.fill_between(monthly_sales.index, monthly_sales.values,
                 alpha=0.2, color='#FF6B6B')

plt.tight_layout()
plt.show()
```
