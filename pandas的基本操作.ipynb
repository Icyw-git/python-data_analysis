```python
import numpy as np
import pandas as pd

data = pd.read_csv('stock_day.csv')
```

```python
data.head()
data = data.drop(['ma5', 'ma10', 'ma20', 'v_ma5', 'v_ma10', 'v_ma20'], axis=1)

#还可以写成data=data.drop(columns=['ma5','ma10','ma20','v_ma5','v_ma10','v_ma20'])
data.head()
```

```python
#直接使用行列索引（先列后行）
print(data['open']['2018-02-27'])

#使用loc 只能指定行列索引的名字
print(data.loc['2018-02-22':'2018-02-27', 'open'])

#使用iloc通过索引的下标去获取
#获取前3天的数据，前5列的结果
print(data.iloc[:3, :5])
```

```python
#赋值操作
#1.直接修改原来的值
data['close'] = 1
#或者
data.close = 1
```

```python
#排序操作
#使用df.sort_values(by=,ascending=) 单个键或多个键进行排序
data.sort_values(by='open', ascending=True).head()
```

```python
data.sort_values(by=['open', 'close'], ascending=True)
```

```python
#使用索引排序
#使用df.sort_index()
data.sort_index()
```

```python
#使用series.sort_index()进行排序
data['p_change'].sort_values(ascending=True).head()
```

```python
#算法运算
#1.add()
data.close.add(2)
data['close'].add(1)
print(data.close + 1)

#2.sub()
print(data.low - 10)
data.low.sub(1)
```

```python
#逻辑运算符
data
```

```python
data[data.open > 23]
#多条件
data[(data.open > 23) & (data.open < 24)]

```

```python
#使用query函数 优化代码
data.query('open>23 & open<24')

#固定值的筛选 isin()
data[data.open.isin([23.53, 23.67])]
data.query('open in [23.53,23.67]')
```

```python
#统计函数
#1.sum()
print(data.sum())
print(data.high.sum())
#2.mean()
print(data.mean())
#3.median()
print(data.median())
#4.min(),max()
print(data.min())

#累计求和
print(data.cumsum())


```

```python
#apply自定义运算
#apply(func,axis=)
#定义一个对列，最大值减最小值的函数
data[['open', 'close']].apply(lambda x: x.max() - x.min(), axis=0)  #axis=0,表示对列进行操作
```

```python
#rank()排名函数
df5 = pd.DataFrame({
    '姓名': ['zizi', 'xixi', 'zhizhi', 'xxixxi'],
    '分数': [100, 90, 90, 80]
}

)
df5
```

```python
df5['成绩排名1'] = df5.分数.rank(method='min', ascending=False)
df5['成绩排名2'] = df5.分数.rank(method='max', ascending=False)
df5['成绩排名3'] = df5.分数.rank(method='dense', ascending=False)
df5['成绩排名4'] = df5.分数.rank(method='average', ascending=False)

df5
```
